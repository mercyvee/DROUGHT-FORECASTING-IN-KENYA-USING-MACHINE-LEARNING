```python
import pandas as pd
import matplotlib as plt
```

```python
file_path = "C:/Users/Admin/Desktop/IODW.xlsx"
```

```python
raw_df = pd.read_excel(file_path, header=None)
```

```python
# Split the single text column by whitespace
df = raw_df[0].str.split(expand=True)
```

```python
df.columns = ['YR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 
              'JUN', 'JUL', 'AUG', 'SEP','OCT', 'NOV', 'DEC']

# Convert to appropriate data types
# Use errors='coerce' to handle any conversion issues
df = df.apply(pd.to_numeric, errors='coerce')

# Preview the cleaned dataframe
df.head()
```

```python
# Save the cleaned dataframe back to the Excel file, replacing the original data
df.to_excel(file_path, index=False)

# Preview the cleaned dataframe
df.head()
```
