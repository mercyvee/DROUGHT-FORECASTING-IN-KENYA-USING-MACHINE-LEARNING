```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "C:/Users/Admin/Desktop/IODE.xlsx"
df = pd.read_excel(file_path)
```

```python
nino_df = df[['YR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 
              'JUN', 'JUL', 'AUG', 'SEP','OCT', 'NOV', 'DEC']]
```

```python
# Pivot the dataframe to get years as rows and months as columns
pivot_df = nino_df.pivot(index='YR', columns='MON', values='NINO4')
```

```python
month_names = {
    1: 'JAN', 2: 'FEB', 3: 'MAR', 4: 'APR', 
    5: 'MAY', 6: 'JUN', 7: 'JUL', 8: 'AUG',
    9: 'SEP', 10: 'OCT', 11: 'NOV', 12: 'DEC'
}
pivot_df = pivot_df.rename(columns=month_names)
```

```python
output_path = "C:/Users/Admin/Desktop/NINO4.xlsx"
pivot_df.to_excel(output_path, index=False)

# Preview the restructured dataframe
print(pivot_df.head())

print(f"Data successfully restructured and saved to {output_path}")
```

```python
file_path = "C:/Users/Admin/Desktop/IODE.xlsx"
ds = pd.read_excel(file_path)
```

```python
plt.figure(figsize=(14, 8))
plt.plot(ds['YR'], ds['SEASON'], marker='o', linestyle='-', color='blue')
plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)

# Add horizontal lines at +1 and -1 standard deviations
plt.axhline(y=1, color='darkred', linestyle='--', alpha=0.5)
plt.axhline(y=-1, color='darkblue', linestyle='--', alpha=0.5)

plt.title('IODE OND Season(1991-2020 Baseline)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Anomaly ()', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
```

```python
output_path = "C:/Users/Admin/Desktop/NINOE4.xlsx"
ds.to_excel(output_path, index=False)
print(f"Data with OND totals and standardized values saved to {output_path}")
```
